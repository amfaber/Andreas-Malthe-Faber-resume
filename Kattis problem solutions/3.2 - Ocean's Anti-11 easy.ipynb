```python
#https://open.kattis.com/problems/anti11
#COMPLETED
def f(n):
    if n == 0:
        return 0
    ones = 1
    zeros = 1
    for i in range(2, n+1):
        ones, zeros = (zeros, ones + zeros)
        print(zeros, ones)
    return (ones + zeros) % (10**9 + 7)

ninput = int(input())
ns = []
for i in range(ninput):
    ns.append(int(input()))

[print(f(n)) for n in ns]
```
