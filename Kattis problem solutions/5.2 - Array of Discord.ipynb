```python
#https://open.kattis.com/problems/arrayofdiscord
#COMPLETED
input()
ints = [i for i in input().split(" ")]
def func(ints):
    for ind, (i, j) in enumerate(zip(ints[:-1], ints[1:])):
        if not len(i) == len(j):
            continue
        if i == "0":
            if j == "9":
                continue
        
        if i[0] == "1" or i[0] == "0":
            if j[0] == "9":
                if len(i) == 1:
                    return (ind+1, "0")
                if int(i[1:]) > int(j[1:]):
                    return (ind+1, "1" + j[1:])
                else:
                    continue
            else:
                return (ind, "9" + i[1:])
        else:
            return (ind+1, "1" + j[1:])
    return None
ans = func(ints)
if ans is None:
    print("impossible")
else:
    out = ints.copy()
    out[ans[0]] = ans[1]
    print(" ".join(out))

```
