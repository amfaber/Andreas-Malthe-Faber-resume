```python
#https://open.kattis.com/problems/abinitio
#IN PROGRESS
def loader():
    V, E, Q = [int(i) for i in input().split(" ")]
    thegraph = [[0]*V for i in range(V)]
    for i in range(E):
        inp = [int(j) for j in input().split(" ")]
        thegraph[inp[0]][inp[1]] = 1
    queries = []
    for i in range(Q):
        queries.append([int(j) for j in input().split(" ")])
    return thegraph, queries


def func1(graph):
    V = len(graph) + 1
    for ver in graph:
        ver.append(0)
    graph.append([0]*V)
    
def func2(graph, x, y):
    graph[x][y] = 1

def func3(graph, x):
    V = len(graph)
    graph[x][:] = [0]*V
    for i in range(V):
        graph[i][x] = 0

def func4(graph, x, y):
    graph[x][y] = 0

def func5(graph):
    graph[:] = [[row[j] for row in graph] for j in range(len(graph))]
    
def func6(graph):
    graph[:] = [[int(not entry) for entry in row] for row in graph]
    for i in range(len(graph)):
        graph[i][i] = 0

def designator(graph, q):
    dic = {1: func1, 2: func2, 3: func3, 4: func4, 5: func5, 6: func6}
    dic[q[0]](graph, *q[1:])

def output(graph):
    V = len(graph)
    print(V)
    for i in range(V):
        inds = [j for j, check in enumerate(graph[i]) if check != 0]
        print(sum(graph[i]), sum(ind*7**j for j, ind in enumerate(inds)) % (10**9+7))

thegraph, queries = loader()
for q in queries:
    designator(thegraph, q)
output(thegraph)
```

```python
import numpy as np
V, Q = 100, 500
graph = np.random.randint(0,2, (V, V))
for i in range(len(graph)):
    graph[i, i] = 0

E = 0
for i in range(graph.shape[0]):
    for j in range(graph.shape[1]):
        if graph[i, j]:
            E += 1

```

```python
def genQ(V, Q):
    dic = {
        1: lambda V: "",
        2: lambda V: " ".join(str(s) for s in np.random.choice(np.arange(V), 2, replace = False)),
        3: lambda V: str(np.random.randint(V)),
        4: lambda V: " ".join(str(s) for s in np.random.choice(np.arange(V), 2, replace = False)),
        5: lambda V: "",
        6: lambda V: ""
    }
    out = ""
    for i in range(Q):
        q = np.random.randint(1, 7)
        out += " ".join([str(q), dic[q](V)]) + "\n"
    return out
```

```python
print(genQ(V, Q))
```
