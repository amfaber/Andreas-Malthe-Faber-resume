```python
def reducer(queries, V):
    out = []
    mode = [0, 0]
    def logic(q, mode = [0, 0]):
        q = q.copy()
        if q[0] == 5:
            return None, [not mode[0], mode[1]]
        elif q[0] == 6:
            return None, [mode[0], not mode[1]]
        else:
            if mode[0]:
                q[1:] = q[:0:-1]
            if mode[1]:
                dic = {1:1, 2: 4, 4: 2, 3: 7, 7: 3}
                q[0] = dic[q[0]]
            return q, mode
    
    for q in queries:
        qp, mode = logic(q, mode)
        if not qp is None:
            out.append(qp)
            if qp[0] == 1:
                V += 1
                if mode[1] == 1:
                    out.append([7, V-1])
    if mode[0] == 1:
        out.append([5])
    if mode[1] == 1:
        out.append([6])
    return out
```
